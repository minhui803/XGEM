```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig1=plt.figure(figsize=(10,8))
methods = ['d=1','d=2','d=3','d=4','d=5','d=6','d=7','d=8','d=9','d=10']
plt.rc('font',family="Times New Roman")
for i,interp_method in zip(np.array(range(1,11,1)),methods):
    axe=fig1.add_subplot(2,5,i)
    x_data = np.array(range(1,10,1))/10
    #x1=np.array[0.1,0.5,0.9]
    y_data = np.array(range(1,21,1))
    axe.set_yticks(np.arange(len(y_data)))
    axe.set_xticks(np.arange(len(x_data)))
    axe.set_xticklabels(x_data,rotation=90)
    axe.set_yticklabels(y_data)
    Z=pd.read_csv(str(i)+".csv",header=None)
    axe.set_xlabel('w')
    axe.set_ylabel('λ')
    axe.set_title(str(interp_method),loc='center')
    values=np.array(Z)
    im=axe.imshow(values,cmap='RdBu_r')
plt.tight_layout()
fig1.subplots_adjust(bottom=0.15)
#fig.colorbar(images[0], ax=axs, orientation='horizontal', fraction=.1)
#axe1=fig1.add_subplot(1,1,1)
#fig1.colorbar(im)
position=fig1.add_axes([0.15, 0.05, 0.7, 0.03])#位置[左,下,右,上]
cb=plt.colorbar(im,cax=position,orientation='horizontal')
#plt.show()
plt.savefig("./PseDSSPC.tif",dpi=300)
plt.savefig("./PseDSSPC.svg",dpi=300)
```
